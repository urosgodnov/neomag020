# üìö Part 2: Variables

Storing and Naming Your Data (45 minutes)

## What is a Variable?

A variable is like a **labeled box** that stores data.

````python
name = "Alice"
````

- `name` is the **label** (variable name)
- `"Alice"` is the **content** (value)
- `=` is the **assignment operator** (puts value in box)

Think of it as: "Create a box labeled 'name' and put 'Alice' inside"

## Why Do We Need Variables?

Without variables:

````python
print(5 + 3)
print(5 * 3)
print(5 - 3)
# What if we want to change 5 to 10? Must change everywhere!
````

With variables:

````python
number = 5
print(number + 3)
print(number * 3)
print(number - 3)
# To change, just modify once: number = 10
````

Variables make code **flexible** and **reusable**.

## Creating Variables

````python
# Text (called a "string")
name = "Alice"

# Whole number (called an "integer")
age = 28

# Decimal number (called a "float")
height = 1.65

# True/False value (called a "boolean")
is_student = True
````

> üìù **Note:** Lines starting with `#` are **comments** - Python ignores them. Use them to explain your code!

## Using Variables

Once created, use variables by their name:

````python
name = "Alice"
age = 28

print(name)
print(age)
print(name)  # Can use as many times as you want
````

**Output:**
```

Alice
```

Variables remember their values until you change them or the program ends.

## Changing Variable Values

Variables can change (that's why they're called "variables"!):

````python
score = 100
print(score)

score = 150
print(score)

score = score + 50  # Use old value to calculate new value
print(score)
````

**Output:**
```

```

## Understanding Assignment

The `=` sign in Python is NOT like math equality!

````python
x = 5      # "Put 5 into the box labeled x"
x = x + 1  # "Take value from x, add 1, put result back in x"
print(x)   # 6
````

This is different from math where `x = x + 1` makes no sense!

In Python, `=` means "assign the value on the right to the variable on the left."

## Variable Naming Rules (MUST Follow)

**Required rules:**

1. Must start with letter or underscore (`_`)
2. Can only contain letters, numbers, underscores
3. Cannot contain spaces or special characters
4. Cannot be Python keywords
5. Case-sensitive (`age` ‚â† `Age` ‚â† `AGE`)

````python
# VALID names:
name = "Alice"
first_name = "Alice"
name2 = "Alice"
_private = "Alice"

# INVALID names (will cause errors):
# 2name = "Alice"      # Starts with number
# first-name = "Alice" # Contains hyphen
# first name = "Alice" # Contains space
# class = "Alice"      # Python keyword
````

## Variable Naming Conventions (SHOULD Follow)

**Python Style Guide (PEP 8) recommendations:**

````python
# GOOD - snake_case (words separated by underscores)
student_name = "Alice"
total_score = 95
is_valid = True

# AVOID - but technically valid
studentName = "Alice"   # camelCase (used in JavaScript)
StudentName = "Alice"   # PascalCase (used for classes)
TOTAL_SCORE = 95        # SCREAMING_CASE (used for constants)

# BAD - hard to read
studentname = "Alice"
s = "Alice"
x = 95
````

## Basic Math Examples

````python
a = 10
b = 3

print(f"Addition:       {a} + {b} = {a + b}")
print(f"Subtraction:    {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division:       {a} / {b} = {a / b:.2f}")
````

**Output:**
```

```

## Division Types Explained

````python
print("Regular division (/):")
print(f"  10 / 3 = {10 / 3}")   # 3.333...
print(f"  10 / 2 = {10 / 2}")   # 5.0 (still float!)

print("\nFloor division (//):")
print(f"  10 // 3 = {10 // 3}") # 3 (rounds down)
print(f"  10 // 2 = {10 // 2}") # 5

print("\nModulo (%):")
print(f"  10 % 3 = {10 % 3}")   # 1 (remainder)
print(f"  10 % 2 = {10 % 2}")   # 0 (evenly divisible)
````

## Practical Uses of Modulo

````python
# Check if number is even or odd
number = 7
if number % 2 == 0:
    print("Even")
else:
    print("Odd")

# Get last digit of a number
year = 2024
last_digit = year % 10
print(f"Last digit of {year} is {last_digit}")  # 4

# Convert minutes to hours and minutes
total_minutes = 135
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"{total_minutes} minutes = {hours}h {minutes}m")
````

## Exponents and Roots

````python
# Powers
print(2 ** 3)     # 8 (2¬≥)
print(10 ** 2)    # 100 (10¬≤)
print(5 ** 0)     # 1 (anything‚Å∞ = 1)

# Square root (power of 0.5)
print(16 ** 0.5)  # 4.0
print(25 ** 0.5)  # 5.0

# Cube root (power of 1/3)
print(27 ** (1/3))  # 3.0

# Scientific notation
distance_to_moon = 3.84e8  # 384,000,000 meters
print(distance_to_moon)
````

## Order of Operations (PEMDAS/BODMAS)

Python follows standard math order:

1. **P**arentheses / **B**rackets
2. **E**xponents / **O**rders
3. **M**ultiplication and **D**ivision (left to right)
4. **A**ddition and **S**ubtraction (left to right)

````python
print(2 + 3 * 4)       # 14 (not 20!)
print((2 + 3) * 4)     # 20
print(2 ** 3 ** 2)     # 512 (exponents: right to left)
print((2 ** 3) ** 2)   # 64
print(10 - 4 - 2)      # 4 (left to right)
````

## Assignment Operators (Shortcuts)

Instead of `x = x + 5`, use shortcuts:

````python
x = 10
print(f"Start: {x}")

x += 5   # Same as: x = x + 5
print(f"After += 5: {x}")

x -= 3   # Same as: x = x - 3
print(f"After -= 3: {x}")

x *= 2   # Same as: x = x * 2
print(f"After *= 2: {x}")

x //= 4  # Same as: x = x // 4
print(f"After //= 4: {x}")
````

## All Assignment Operators

| `*=` | `x *= 5` | `x = x * 5` |

| `%=` | `x %= 5` | `x = x % 5` |
| `**=` | `x **= 5` | `x = x ** 5` |

# üìö Part 5: F-Strings

Formatted String Literals (20 minutes)

## The Problem

How do we combine text and variables?

````python
name = "Alice"
age = 28

# Concatenation is clunky:
message = "My name is " + name + " and I am " + str(age) + " years old."
print(message)
````

There must be a better way!

## F-Strings to the Rescue!

Put `f` before the string and variables inside `{curly braces}`:

````python
name = "Alice"
age = 28

# F-string way - clean and readable!
message = f"My name is {name} and I am {age} years old."
print(message)
````

**Output:**
```
My name is Alice and I am 28 years old.
```

## F-String Syntax

````python
name = "Alice"
score = 95

# The f before quotes is required!
greeting = f"Hello, {name}!"
result = f"You scored {score} points."

print(greeting)
print(result)

# Without the f, braces are literal text:
no_f = "Hello, {name}!"  # Prints: Hello, {name}!
print(no_f)
````

## Expressions in F-Strings

You can put any Python expression inside the braces:

````python
x = 10
y = 5

print(f"Sum: {x + y}")
print(f"Product: {x * y}")
print(f"x squared: {x ** 2}")
print(f"Is x bigger? {x > y}")
print(f"Division: {x / y}")
````

**Output:**
```
Sum: 15
Product: 50
x squared: 100
Is x bigger? True
Division: 2.0
```

## Function Calls in F-Strings

````python
name = "alice"
items = ["apple", "banana", "cherry"]

print(f"Original: {name}")
print(f"Upper: {name.upper()}")
print(f"Title: {name.title()}")
print(f"Length: {len(name)}")
print(f"Number of items: {len(items)}")
````

**Output:**
```
Original: alice
Upper: ALICE
Title: Alice
Length: 5
Number of items: 3
```

## Number Formatting: Decimal Places

````python
pi = 3.14159265359
price = 19.5

# .Xf = X decimal places
print(f"Pi: {pi}")           # 3.14159265359
print(f"Pi: {pi:.2f}")       # 3.14
print(f"Pi: {pi:.4f}")       # 3.1416

print(f"Price: ${price}")    # $19.5
print(f"Price: ${price:.2f}")# $19.50
````

The `f` after the number means "fixed-point" (show as decimal).

## Number Formatting: Thousands Separator

````python
population = 2100000
salary = 52750.50

# Use comma for thousands separator
print(f"Population: {population:,}")
print(f"Salary: ${salary:,.2f}")

# Use underscore (Python 3.6+)
print(f"Population: {population:_}")
````


## Number Formatting: Percentages

````python
ratio = 0.756
accuracy = 0.9823
completion = 0.5

print(f"Ratio: {ratio:.1%}")       # 75.6%
print(f"Accuracy: {accuracy:.2%}") # 98.23%
print(f"Complete: {completion:.0%}")# 50%
````

The `%` multiplies by 100 and adds the percent sign.

## Number Formatting: Alignment and Padding

````python
name = "Alice"
score = 42

# Width and alignment
print(f"|{name:>15}|")  # Right align, width 15
print(f"|{name:<15}|")  # Left align
print(f"|{name:^15}|")  # Center

# Zero padding
print(f"ID: {score:05}")    # 00042
print(f"Hex: {score:04x}")  # 002a
````

**Output:**
```
|          Alice|
|Alice          |
|     Alice     |
ID: 00042
Hex: 002a
```

## Day of the Week

````python
from datetime import date

today = date.today()

# weekday() returns 0-6 (Monday=0, Sunday=6)
print(f"Weekday number: {today.weekday()}")

# Get the name
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
        'Friday', 'Saturday', 'Sunday']
print(f"Day name: {days[today.weekday()]}")
````

## Date Arithmetic

````python
from datetime import date, timedelta

today = date.today()

# Add days
tomorrow = today + timedelta(days=1)
next_week = today + timedelta(weeks=1)
print(f"Tomorrow: {tomorrow}")
print(f"Next week: {next_week}")

# Subtract days
yesterday = today - timedelta(days=1)
print(f"Yesterday: {yesterday}")
````

## Calculating Differences

````python
from datetime import date

# Two dates
start = date(2024, 1, 1)
end = date(2024, 12, 31)

# Difference
difference = end - start
print(f"Days between: {difference.days}")
# Output: Days between: 365
````

## Calculating Age

````python
from datetime import date

# Birth date
birthdate = date(1990, 5, 15)
today = date.today()

# Simple age calculation
age = today.year - birthdate.year

# Adjust if birthday hasn't occurred this year
if (today.month, today.day) < (birthdate.month, birthdate.day):
    age -= 1

print(f"Age: {age} years")
````

```
Traceback (most recent call last):
  File "script.py", line 3, in <module>

NameError: name 'nmae' is not defined
```

1. **Traceback** - The path to the error
2. **File and line** - Exact location
3. **Code** - The line that caused the error
4. **Error type** - What kind of error (NameError)
5. **Message** - Description of the problem

**Always read the LAST line first!**

````python
name = "Alice"
print(nmae)  # Typo!
````

**Error:**

````python
print("\n" + "=" * 45)
print("SIMPLE INTEREST CALCULATOR".center(45))
print("=" * 45)

principal = float(input("\nPrincipal amount ($): "))
rate = float(input("Annual interest rate (%): ")) / 100
years = int(input("Investment period (years): "))

interest = principal * rate * years
total = principal + interest

print("\n" + "-" * 45)
print("INVESTMENT SUMMARY".center(45))
print("-" * 45)
print(f"Principal Amount:    ${principal:>15,.2f}")
print(f"Interest Rate:       {rate*100:>15.2f}%")
print(f"Time Period:         {years:>15} years")
print("-" * 45)
print(f"Interest Earned:     ${interest:>15,.2f}")
print(f"Total Value:         ${total:>15,.2f}")
print("=" * 45)
print(f"\nüí∞ Your money grew by {(total/principal - 1)*100:.1f}%!")
````
